## UMAP

- `all-MiniLM-L6-v2`
- `DeepPavlov/rubert-base-cased-sentence`
- `sentence-transformers/distiluse-base-multilingual-cased-v1`
- `cointegrated/rubert-tiny2`
- `ai-forever/sbert_large_nlu_ru`

```python
from bertopic import BERTopic
from flair.embeddings import TransformerDocumentEmbeddings, TransformerWordEmbeddings, DocumentPoolEmbeddings

sentence_model_v1 = TransformerWordEmbeddings("DeepPavlov/rubert-base-cased-sentence")
sentence_model_v2 = TransformerWordEmbeddings('sentence-transformers/distiluse-base-multilingual-cased-v1')
sentence_model_v3 = TransformerWordEmbeddings('cointegrated/rubert-tiny2')
```

```python
document_embeddings_v1 = DocumentPoolEmbeddings([sentence_model_v1])
document_embeddings_v2 = DocumentPoolEmbeddings([sentence_model_v2])
document_embeddings_v3 = DocumentPoolEmbeddings([sentence_model_v3])
```

```python
# Reduce dimensionality
umap_model = UMAP(n_neighbors=15,
                  n_components=5,
                  min_dist=0.0,
                  metric='cosine',
                  random_state=100)
```

```python
# Cluster reduced embeddings
hdbscan_model = HDBSCAN(min_cluster_size=10, metric='euclidean', cluster_selection_method='eom', prediction_data=True)

# topic_model = BERTopic(hdbscan_model=hdbscan_model)
```

```python
# Tokenize topics
vectorizer_model = CountVectorizer()
```

```python
# Create topic representation
ctfidf_model = ClassTfidfTransformer()
```

### 1 Модель с первой model embeddings

```python
%%time

topic_model_v1 = BERTopic(
    embedding_model = document_embeddings_v1,
    umap_model = umap_model,
    hdbscan_model = hdbscan_model,
    vectorizer_model = vectorizer_model,
    ctfidf_model = ctfidf_model,
    n_gram_range=(1,3),
    calculate_probabilities=True,
    language="russian"
)

topics_v1, probabilities_v1 = topic_model_v1.fit_transform(df['cleaned_answers_lemm'])
```
